```python
import numpy as np
import pandas as pd
import nibabel as nb
import matplotlib.pyplot as plt
import os
from nilearn import surface
%matplotlib inline
```

```python
path = "/home/brunovieira/backup/Tamires_Experiments/Data/BigBrain/"
lh_thickness_df = pd.DataFrame(np.load(path+'thicknesses_left.npy')).T
lh_thickness_df = lh_thickness_df.rename(columns={0: "thickness_layer1", 
                                            1: "thickness_layer2",
                                            2: "thickness_layer3",
                                            3: "thickness_layer4",
                                            4: "thickness_layer5",
                                            5: "thickness_layer6",
                                            6: "total_thickness",
                                            })
lh_thickness_df["BN_atlas"] = surface.load_surf_data(path+"lh.BN_atlas.annot")
lh_thickness_df["Economo_atlas"] = surface.load_surf_data(path+"lh.economo.annot")
lh_thickness_df["DKT_atlas"] = surface.load_surf_data(path+"lh.DKTatlas40.annot")
lh_thickness_df["HCP_atlas"] = surface.load_surf_data(path+"lh.HCP-MMP-1.0.annot")
lh_thickness_df
```

```python
lh_thickness_df['percentage_layer1'] = lh_thickness_df.thickness_layer1/lh_thickness_df.total_thickness
lh_thickness_df['percentage_layer2'] = lh_thickness_df.thickness_layer2/lh_thickness_df.total_thickness
lh_thickness_df['percentage_layer3'] = lh_thickness_df.thickness_layer3/lh_thickness_df.total_thickness
lh_thickness_df['percentage_layer4'] = lh_thickness_df.thickness_layer4/lh_thickness_df.total_thickness
lh_thickness_df['percentage_layer5'] = lh_thickness_df.thickness_layer5/lh_thickness_df.total_thickness
lh_thickness_df['percentage_layer6'] = lh_thickness_df.thickness_layer6/lh_thickness_df.total_thickness
lh_thickness_df
```

```python
layers_BN = lh_thickness_df.astype('f8').groupby(by="BN_atlas").mean()
layers_BN = layers_BN.drop(columns=["Economo_atlas","DKT_atlas", "HCP_atlas"])
# duplicando dados para hemisferio direito
layers_BN_ = layers_BN.copy()
layers_BN_.index = layers_BN_.index + 1
layers_BN_ = layers_BN_.drop(1, axis=0)
layers_BN = pd.concat([layers_BN, layers_BN_]).sort_index()
layers_BN
#layers_BN.to_csv(path+"layers_BN.csv")
```

```python
layers_Economo = lh_thickness_df.astype('f8').groupby(by="Economo_atlas").mean()
layers_Economo = layers_Economo.drop(columns=["BN_atlas","DKT_atlas", "HCP_atlas"])
layers_Economo
#layers_Economo.to_csv(path+"layers_Economo.csv")
```
